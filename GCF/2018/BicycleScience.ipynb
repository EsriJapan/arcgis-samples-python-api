## GIS 解析を実行する：「エリア集計」で東京都 23 区ごとの放置自転車数を解析する

### 「エリア集計」解析とは

![エリア集計解析](data/summerizeWithin.png "サンプル")

### 境界データを定義する（東京都のポリゴンデータ）

```python
tokyo_polygon = {"type":"FeatureLayer",
                "url":"https://services3.arcgis.com/iH4Iz7CEdh5xTJYb/arcgis/rest/services/CITY/FeatureServer/0",
                "opacity":0.75}
# 表示している地図に重ねる
DataScienceMap.add_layer(tokyo_polygon)
```

__[地図の確認](#準備：Jupyter-Notebookへ地図を表示させる)__

### 「エリア集計」解析を実行する

```python
from arcgis.features import summarize_data
summarize_layer = summarize_data.summarize_within(tokyo_polygon,
                                illegal_parking_bicycles_layer,
                                sum_shape=True,
                                shape_units='SquareMiles',
                                summary_fields=["放置台数_自転車 SUM"],
                                output_name='東京都エリア集計_2015',
                                gis=gis)
summarize_layer
```

```python
# シンボルの変更
from arcgis import GIS, features
import re

serviceIndex = r'/0'
# シンボル更新レイヤー
update_summarize_layer = summarize_layer.layers[0].url
adminURL = re.sub(r'(/rest/)', r'\1admin/', update_summarize_layer)
editfl = features.managers.FeatureLayerManager(adminURL, gis)

# シンボル定義取得:すでに定義しているレイヤーから呼び出して適用する
summarize_layer_symboled = gis.content.get('c70119484a5240d8af8fa92e32cc21d2')
drawingInfo = summarize_layer_symboled.layers[0].properties['drawingInfo']
draInf = {}
draInf['drawingInfo'] = drawingInfo

# シンボルの更新
updatesymbol = editfl.update_definition(draInf)
```

```python
# 解析結果を地図へ表示する
DataScienceMap.add_layer(summarize_layer)
```

### 3.結果をグラフ出力

```python
import io
import pandas
from IPython.display import display
import json
```

```python
layer = summarize_layer.layers[0]
layer.query().df
```

### 必要な要素だけでまとめる

```python
illegal_parking_bicycles_df = layer.query().df
# 3つの要素だけで表示
filtered_df = illegal_parking_bicycles_df[['CSS_NAME', 'SUM_放置台数_自転車','Point_Count']]
# 要素でソートして表示
sorted_df = filtered_df.sort_values(by='SUM_放置台数_自転車',ascending=False)
sorted_df
```

### グラフ出力

```python
import matplotlib.pyplot as pd
from matplotlib.font_manager import FontProperties
# matplotlibはデフォルトで日本語フォントをもっていないそうなので、豆腐（文字化け）の場合はフォントを指定します。
fp = FontProperties(fname=r'data/ipaexg.ttf', size=11)
# グラフの設定
graphdata = sorted_df.plot(x='CSS_NAME', y='SUM_放置台数_自転車', kind='bar', figsize=(10,6))
graphdata.set_xticklabels(sorted_df['CSS_NAME'], fontproperties=fp)
graphdata.set_xlabel('CSS_NAME', fontproperties=fp)
pd.show()
```
